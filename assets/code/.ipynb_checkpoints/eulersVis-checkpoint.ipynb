```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-muted')
params = {'axes.labelsize': 20,
          'axes.titlesize': 20}
plt.rcParams.update(params)
import numpy as np
import pandas as pd
```

```python
TRUE_VALUE = np.e - 2

def yPrime(t, y):
    return t + y

def eulersMethodAll(yPrime, N, a, b, alpha):
    h = (b-a) / N
    w = alpha
    times = [a]
    results = [w]
    for i in range(N+1):
        time = a + i * h
        times.append(time)
        w = w + h * yPrime(time, w)
        results.append(w)
    return (times,results)
```

```python
demo = eulersMethodAll(yPrime, N=4, a=0, b=1, alpha=0)
```

```python
f, ax = plt.subplots(figsize=(13,7))
colors =  plt.get_cmap('Set2')

true_y = [np.exp(i) - i - 1 for i in np.arange(0,1,0.001)]

ax.plot(np.arange(0,1,0.001), true_y, color=colors(1), label='true')
ax.plot(demo[0], demo[1], color=colors(2))
plt.legend()
# plt.savefig('f1.png', bbox_inches='tight')
plt.show()
```
